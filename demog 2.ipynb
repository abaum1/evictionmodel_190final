```python
data_path = '/Users/ameliabaum/Desktop/Amelia/Berkeley 2018-2019/EnRes C190/evictionmodel_190final/data/'
#table B02001 race
#race_feature = {}
race_feature = []
for race_file in os.listdir(data_path+'B02001_race'):
    year = '20'+race_file.split("_")[1]
    #if year in ['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016']:
    if year in ['2017']:

        df = pd.read_csv(data_path+'B02001_race/'+race_file)
        df["Year"] = year    
        race_feature.append(df)
allyears_race = pd.concat(race_feature, axis=0, sort=False)
allyears_race["geoid_year"] = allyears_race['Id2'].astype(str) + allyears_race["Year"].astype(str)


#table B25071 rent burden
burden_feature = []
for burden_file in os.listdir(data_path+'B25071_burden'):
    year = '20'+burden_file.split("_")[1]
    #if year in ['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016']:
    if year in ['2017']:
        df = pd.read_csv(data_path+'B25071_burden/'+burden_file)
        df["Year"] = year
        
        burden_feature.append(df)
allyears_burden = pd.concat(burden_feature, axis=0, sort=False)
allyears_burden["geoid_year"] = allyears_burden['Id2'].astype(str) + allyears_burden["Year"].astype(str)        


#table S1101 household size
hsize_feature = []
for size_file in os.listdir(data_path+'S1101_size'):
    year = '20'+size_file.split("_")[1]
   #if year in ['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016']:
    if year in ['2017']:
        df = pd.read_csv(data_path+'S1101_size/'+size_file)
        df["Year"] = year
        
        hsize_feature.append(df)
allyears_hsize = pd.concat(hsize_feature, axis=0, sort=False)
allyears_hsize["geoid_year"] = allyears_hsize['Id2'].astype(str) + allyears_hsize["Year"].astype(str)

#table S1501 % with bachelors degree
edu_feature = []
for edu_file in os.listdir(data_path+'S1501_edu'):
    year = '20'+edu_file.split("_")[1]
    #if year in ['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016']:
    if year in ['2017']:
        df = pd.read_csv(data_path+'S1501_edu/'+edu_file)
        df["Year"] = year
        
        edu_feature.append(df)
allyears_edu = pd.concat(edu_feature, axis=0, sort=False)
allyears_edu["geoid_year"] = allyears_edu['Id2'].astype(str) + allyears_edu["Year"].astype(str)
# allyears_built = allyears_built.rename(columns={"Id2":"Geoid", "Geography": "Census Tract", 'Estimate; Median year structure built': "Median year structure built", 
#         "Year": "Year", "geoid_year": "geoid_year"})


#table B25035 median year built
built_feature = []
for built_file in os.listdir(data_path+'B25035_yearbuilt'):
    year = '20'+built_file.split("_")[1]
    #if year in ['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016']:
    if year in ['2017']:
        df = pd.read_csv(data_path+'B25035_yearbuilt/'+built_file)
        df["Year"] = year
        built_feature.append(df)
allyears_built = pd.concat(built_feature, axis=0, sort=False)
allyears_built["geoid_year"] = allyears_built['Id2'].astype(str) + allyears_built["Year"].astype(str)
# allyears_built = allyears_built.rename(columns={"Id2":"Geoid", "Geography": "Census Tract", 'Estimate; Median year structure built': "Median year structure built", 
#         "Year": "Year", "geoid_year": "geoid_year"})



#table B25039 median renter moved in
moved_feature = []
for moved_file in os.listdir(data_path+'B25039_moved'):
    year = '20'+ moved_file.split("_")[1]
    #if year in ['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016']:
    if year in ['2017']:
        df = pd.read_csv(data_path+'B25039_moved/'+moved_file)
        df["Year"] = year
        moved_feature.append(df)
        print(df.columns)
allyears_moved = pd.concat(moved_feature, axis=0, sort=False)
allyears_moved["geoid_year"] = allyears_moved['Id2'].astype(str) + allyears_moved["Year"].astype(str)
allyears_moved_rename = allyears_moved.rename(columns={"Id2":"Geoid", "Geography": "Census Tract", 
'Estimate; Renter occupied': "Median Year Renter Moved Into Unit", 
'Estimate; Median year householder moved into unit -- - Renter occupied':"Median Year Renter Moved Into Unit"})
#.drop("drop this", axis=1)


#table B25001 housing units in tract
units_feature = []
for units_file in os.listdir(data_path+'B25001_units'):
    year = '20'+units_file.split("_")[1]
    #if year in ['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016']:
    if year in ['2017']:
        df = pd.read_csv(data_path+'B25001_units/'+units_file)
        df["Year"] = year
        
        units_feature.append(df)
allyears_units = pd.concat(units_feature, axis=0, sort=False)
allyears_units["geoid_year"] = allyears_units['Id2'].astype(str) + allyears_units["Year"].astype(str)
allyears_units.rename(columns={"Id2":"Geoid", "Geography": "Census Tract", "Estimate; Total": "Housing Units", 
                       "Year": "Year", "geoid_year": "geoid_year"}).head()
```

```python
#grouped_evictions = pd.read_csv('/Users/ameliabaum/Desktop/Amelia/Berkeley 2018-2019/EnRes C190/evictionmodel_190final/data/evictions_geocoded11_tractandyear.csv')
grouped_evictions = pd.read_csv('data/evictions_geocoded17_tractandyear.csv')
grouped_evictions["num evictions"] = grouped_evictions["Owner Move In"]
grouped_evictions["geoid_year"] = (grouped_evictions["Geoid"].astype(str) + grouped_evictions["Year"].astype(str))
grouped_evictions["geoid_year"] = grouped_evictions["geoid_year"].apply(lambda x: x[1:])
grouped_evictions = grouped_evictions[["Geoid", "Year", "geoid_year", "num evictions"]]
grouped_evictions.head()
```

```python
race = grouped_evictions.merge(allyears_race, how='left', left_on="geoid_year", right_on="geoid_year")

race_edu = race.merge(allyears_edu, how='left', left_on="geoid_year", right_on="geoid_year")
race_edu_units = race_edu.merge(allyears_units, how='left', left_on="geoid_year", right_on="geoid_year")
race_edu_units_built = race_edu_units.merge(allyears_built, how='left', left_on="geoid_year", right_on="geoid_year")

race_edu_units_built_moved = race_edu_units_built.merge(allyears_moved, how='left', left_on="geoid_year", right_on="geoid_year")

race_edu_units_built_moved_hsize = race_edu_units_built_moved.merge(allyears_hsize, how='left', left_on="geoid_year", right_on="geoid_year")

race_edu_units_built_moved_burden = race_edu_units_built_moved_hsize.merge(allyears_burden, how='left', left_on="geoid_year", right_on="geoid_year")


```

```python
clean_data = race_edu_units_built_moved_burden[['Geoid','Year_x', 'num evictions', 'geoid_year', 'Estimate; Total:', 'Estimate; Total: - White alone',
       'Estimate; Total: - Black or African American alone',
       'Estimate; Total: - American Indian and Alaska Native alone',
       'Estimate; Total: - Asian alone',
    'Estimate; Median year householder moved into unit -- - Renter occupied',
       'Estimate; Total: - Native Hawaiian and Other Pacific Islander alone', 
        'Percent; Estimate; Population 25 years and over - Bachelor\'s degree', 'Estimate; Total', 'Estimate; Median year structure built', 
        'Total; Estimate; Average household size', 'Estimate; Median gross rent as a percentage of household income']]
clean_data["Year"] = clean_data['Year_x'].iloc[:,0]
clean_data1 = clean_data.drop("Year_x", axis=1)
```

```python
clean_data1 = clean_data1[clean_data1["Geoid"] != "Nan"] #this drops the last 4 rows which is correcct
clean_data1
```

```python
clean_data1["Median Year Structure Built"] = clean_data1["Estimate; Median year structure built"].astype(str).apply(lambda x: x.replace("+", "").replace("-", ""))
clean_data1["Median Year Moved In"] = clean_data1["Estimate; Median year householder moved into unit -- - Renter occupied"].astype(str).apply(lambda x: x.replace("+", "").replace("-", ""))
clean_data1["Percent with Bachelor's degree"] = clean_data1['Percent; Estimate; Population 25 years and over - Bachelor\'s degree'].astype(str).apply(lambda x: x.replace("+", "").replace("-", ""))
clean_data1["Median rent burden"] = clean_data1["Estimate; Median gross rent as a percentage of household income"].astype(str).apply(lambda x: x.replace("+", "").replace("-", ""))

renamed_clean = clean_data1.drop(["Estimate; Median year structure built", "Estimate; Median year householder moved into unit -- - Renter occupied", 'Percent; Estimate; Population 25 years and over - Bachelor\'s degree', "Estimate; Median gross rent as a percentage of household income"], axis=1)
```

```python
empty = renamed_clean["Median Year Moved In"] == ''
renamed_clean["Median Year Moved In"][empty] = 'nan'
renamed_clean["Median Year Moved In"] = renamed_clean["Median Year Moved In"].astype(float)
renamed_clean["Median Year Moved In"].dtype
```

```python
renamed_clean = clean_data1.rename({"Estimate; Total:": "Total Population", 
                                   "Estimate; Total": "Total Housing Units"}, axis=1)
renamed_clean["Percent with Bachelor\'s degree"] = renamed_clean["Percent with Bachelor\'s degree"].astype(float)
renamed_clean["Total; Estimate; Average household size"] = renamed_clean["Total; Estimate; Average household size"].astype(float)

#renamed_clean["Estimate; Median gross rent as a percentage of household income"] = renamed_clean["Estimate; Median gross rent as a percentage of household income"].astype(float)

# renamed_clean["Median Year Structure Built"] = renamed_clean["Median Year Structure Built"].astype(float)
empty_mysb = renamed_clean["Median Year Structure Built"] == ''
renamed_clean["Median Year Structure Built"][empty_mysb] = 'nan'
renamed_clean["Median Year Structure Built"] = renamed_clean["Median Year Structure Built"].astype(float)


# renamed_clean["Median Year Moved In"] = renamed_clean["Median Year Moved In"].astype(float)
empty_mymi = renamed_clean["Median Year Moved In"] == ''
renamed_clean["Median Year Moved In"][empty_mymi] = 'nan'
renamed_clean["Median Year Moved In"] = renamed_clean["Median Year Moved In"].astype(float)

empty_mrb = renamed_clean["Median rent burden"] == ''
renamed_clean["Median rent burden"][empty_mrb] = 'nan'
renamed_clean["Median rent burden"] = renamed_clean["Median rent burden"].astype(float)

renamed_clean = renamed_clean.drop(["Estimate; Median year householder moved into unit -- - Renter occupied", "Estimate; Median year structure built"], axis=1)
renamed_clean["Evictions per units"] = renamed_clean["num evictions"]/renamed_clean["Total Housing Units"]
renamed_clean.dtypes

renamed_clean = renamed_clean.drop(["Estimate; Median gross rent as a percentage of household income", "Percent; Estimate; Population 25 years and over - Bachelor's degree"], axis=1)
renamed_clean.dtypes
```
