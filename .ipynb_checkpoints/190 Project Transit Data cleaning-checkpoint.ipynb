## Muni bus stop data

```python
muni_path = '/Users/ameliabaum/Desktop/Amelia/Berkeley 2018-2019/EnRes C190/evictionmodel_190final/google_transit/'
stops = pd.read_csv(muni_path+ 'stops.txt')
shapes = pd.read_csv(muni_path+ 'shapes.txt')
routes = pd.read_csv(muni_path+ 'routes.txt')
stop_times = pd.read_csv(muni_path + 'stop_times.txt')
trips = pd.read_csv(muni_path +'trips.txt')
```

```python
trips_and_shapes = trips.merge(shapes, how='inner', left_on='shape_id', right_on='shape_id')
trips_shapes_routes = trips_and_shapes.merge(routes, how='left', left_on='route_id', right_on='route_id')
#len(trips_shapes_routes) #try filtering by R and X- these are rapid and express routes
```

```python
trips_shapes_routes_X = trips_shapes_routes[trips_shapes_routes["route_long_name"].str.contains("X")]
trips_shapes_routes_XR = trips_shapes_routes_X[trips_shapes_routes_X["route_long_name"].str.contains("R")]
len(trips_shapes_routes_XR)
```

# Ping FCC API to geocode census tracts


```python
def getGeoid(lon, lat):
#pings the FCC API to geocode lat lon coordinates to other geograpies.
    try:
        result = cg.coordinates(lon, lat)
        geoid = result['Census Tracts'][0]['GEOID']
        return geoid
    except(ValueError, TypeError):
        return 'Nan'
```

```python
start = time.time()
stops["Geoid"] = stops.apply(lambda row: getGeoid(row["stop_lon"], row["stop_lat"]), axis=1)

end = time.time()
elapsed = end-start #should take about an hour
elapsed
```

```python
stops.to_csv("data/geocoded_muni_stops.csv")
```

```python
stops_grouped = stops.groupby("Geoid").count()
stops_grouped["num stops"] = stops_grouped["stop_id"]
stops_grouped = stops_grouped["num stops"]
```

# Group by Census Tract to count stops per tract

```python
stops_grouped.to_csv("data/geocoded_muni_stops_bytract.csv")
```

## Commutter Shuttle Program Data

```python
shuttle_path = '/Users/ameliabaum/Desktop/Amelia/Berkeley 2018-2019/EnRes C190/evictionmodel_190final/MTA.speedbumps/'

commuter_program_shuttles = gp.read_file(shuttle_path+ 'geo_export_e2840cb2-6482-4a2f-bd56-0878a2cbccf3.shp')
len(commuter_program_shuttles)
```

```python
commuter_program_shuttles["Location"] = commuter_program_shuttles["geometry"].astype(str).apply(lambda row: row.split("(")[1])
commuter_program_shuttles["Latitude"] = commuter_program_shuttles["Location"].astype(str).apply(lambda row: row.split(" ")[1][:-1])
commuter_program_shuttles["Longitude"] = commuter_program_shuttles["Location"].astype(str).apply(lambda row: row.split(" ")[0])
```

```python
start = time.time()
commuter_program_shuttles["Geoid"] = commuter_program_shuttles.apply(lambda row: getGeoid(row["Longitude"], row["Latitude"]), axis=1)
# foo = commuter_program_shuttles.head(10).apply(lambda row: getGeoid(row["Longitude"], row["Latitude"]), axis=1)

end = time.time()
elapsed = end-start #should take about a minute
elapsed
```

```python
commuter_program_shuttles = commuter_program_shuttles[['Geoid','from_st', 'to_st', 'geometry', 'Latitude', 'Longitude']]
commuter_program_shuttles.to_csv("/Users/ameliabaum/Desktop/Amelia/Berkeley 2018-2019/EnRes C190/evictionmodel_190final/data/shuttles_geocodedbytract.csv")



```
